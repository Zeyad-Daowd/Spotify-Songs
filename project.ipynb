```python
all_genres = dict()
for i in range(len(spotify_data['Genre'])):
    intermediate =spotify_data['Genre'][i].split("'") 
    if '[' in intermediate[:]:
        intermediate.remove('[')
    if ']' in intermediate[:]:
        intermediate.remove(']')
    if '[]' in intermediate[:]:
        intermediate.remove('[]')
    if ', ' in intermediate[:]:
        intermediate = list(filter(lambda a : a != ', ',intermediate))
    for word in intermediate:
        if word in all_genres:
            all_genres[word] = all_genres[word] + 1
        else:
            all_genres[word] = 1
```

```python
from itertools import islice

def take(n, my_dict):
    """Return the first n items of the iterable as a list."""
    return list(islice(my_dict, n))
```

```python
sorted_tuples =sorted(all_genres.items(), key=lambda x:x[1],reverse=True) 
sorted_dict = dict(sorted_tuples)
top20_genres = take(20, sorted_dict)
sorted_data = []
for i in top20_genres:
    sorted_data.append(all_genres[i])
plt.bar(top20_genres, sorted_data)
plt.xticks(rotation = 'vertical')
```

```python
genre_names = take(1545, all_genres)
streams_on_genre = dict()
for genre in genre_names:
    streams = 0
    for index, row in spotify_data.iterrows():
        if genre in row['Genre']:
            streams += row['Streams']
    streams_on_genre[genre] = streams
```

```python
def map_chords(x):
    if x == "G#/Ab":
        return 1
    elif x == "A":
        return 2
    elif x == "D#/Eb":
        return 3
    elif x == "F#/Gb":
        return 4
    elif x == "E":
        return 5
    elif x == "F":
        return 6
    elif x == "B":
        return 7
    elif x == "C#/Db":
        return 8
    elif x == "A#/Bb":
        return 9
    elif x == "G":
        return 10
    elif x == "D":
        return 11
    else:
        return 12
```

```python
def plot_line(x,y,axes,j,a,b):
    slope, intercept = np.polyfit(x, y, 1)

    # Generate y values for the regression line
    regression_line = slope * x + intercept

    # Plot the data points
    axes[j].scatter(x, y, label='Data Points')

    # Plot the regression line
    axes[j].plot(x, regression_line, color='red', label='Regression Line')

    # Add labels and title
    axes[j].set_xlabel(a)
    axes[j].set_ylabel(b[j])
    axes[j].set_title(a + " vs " + b[j])

    # Add a legend
    axes[j].legend()

    
```

### Question # : Is there a relation between the genre of a song and [Valence, Energy, Temp]?

```python
total_energy = []
total_temp = []
total_valence = []
for genre in sorted_dict2:
    energy_ = 0
    tempo_ = 0
    valence_ = 0
    for index, row in spotify_data.iterrows():
        if genre in row['Genre']:
            energy_ += row['Energy']
            tempo_ += row['Tempo']
            valence_ += row['Valence']
    total_energy.append(energy_)
    total_temp.append(tempo_)
    total_valence.append(valence_)
```

```python
# get average for all
total_energy = list(np.array(total_energy) / len(total_energy))  
total_valence = list(np.array(total_valence) / len(total_valence))
total_temp = list(np.array(total_temp) / len(total_temp))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
features = spotify_data[['Highest Charting Position', 'Number of Times Charted', 'Popularity',
                         'Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness',
                         'Liveness', 'Duration (ms)', 'Valence', 'Chord', 'Genre_mapped',
                         'Temp_mapped', 'Artist_Followers']]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Plot the predicted vs. actual values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Streams')
plt.ylabel('Predicted Streams')
plt.title('Actual vs. Predicted Streams (Linear Regression)')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

degree = 2

poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

model = LinearRegression()

model.fit(X_train_poly, y_train)

y_pred = model.predict(X_test_poly)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

plt.scatter(y_test, y_pred)
plt.xlabel('Actual Streams')
plt.ylabel('Predicted Streams')
plt.title('Actual vs. Predicted Streams (Polynomial Regression)')

plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor



X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

decision_tree_model = DecisionTreeRegressor(random_state=42)
decision_tree_model.fit(X_train, y_train)

y_pred_decision_tree = decision_tree_model.predict(X_test)

mse_decision_tree = mean_squared_error(y_test, y_pred_decision_tree)
r2_decision_tree = r2_score(y_test, y_pred_decision_tree)

print('Decision Tree Regressor:')
print(f'Mean Squared Error: {mse_decision_tree}')
print(f'R-squared: {r2_decision_tree}')

plt.scatter(y_test, y_pred_decision_tree)
plt.xlabel('Actual Streams')
plt.ylabel('Predicted Streams')
plt.title('Actual vs. Predicted Streams (Decision Tree Regression)')

plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor



X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

random_forest_model = RandomForestRegressor(random_state=42)
random_forest_model.fit(X_train, y_train)

y_pred_random_forest = random_forest_model.predict(X_test)

mse_random_forest = mean_squared_error(y_test, y_pred_random_forest)
r2_random_forest = r2_score(y_test, y_pred_random_forest)

print('Random Forest Regressor:')
print(f'Mean Squared Error: {mse_random_forest}')
print(f'R-squared: {r2_random_forest}')

plt.scatter(y_test, y_pred_random_forest)
plt.xlabel('Actual Streams')
plt.ylabel('Predicted Streams')
plt.title('Actual vs. Predicted Streams (Random Forest Regression)')

plt.show()
```
